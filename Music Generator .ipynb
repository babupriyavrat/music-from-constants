```python
from pprint import pprint

def get_base_frequencies():
    base_frequenices=[]
    for i in range(0,9):
        base_freq=16.352*pow(2,i)
        base_frequenices.append(base_freq)
    return base_frequenices

base_freq=get_base_frequencies()
print(base_freq)

def get_piano_notes(base_freq,base):
    '''
    Returns a dict object for all the piano 
    note's frequencies
    '''
    # White keys are in Uppercase and black keys (sharps) are in lowercase
    octave = ['C'+str(base), 'c'+str(base), 'D'+str(base), 'd'+str(base), 'E'+str(base), \
              'F'+str(base), 'f'+str(base), 'G'+str(base), 'g'+str(base), 'A'+str(base), 'a'+str(base), 'B'+str(base)] 
    
    note_freqs = {octave[i]: base_freq * pow(2,(i/12)) for i in range(len(octave))}        
    note_freqs[''] = 0.0 # silent note
    
    return note_freqs
  
note_matrix={}
for i in range(0,9):
    note_matrix.update(get_piano_notes(base_freq[i],i))
    
    
# To get the piano note's frequencies
#note_freqs_1 = get_piano_notes()
pprint(note_matrix)
```

```python
#some irrational number - sin
min_octave=3
max_octave=5
def convert_decimal_to_notes(sin):
    whole,decimals= sin.split('.')
    current_octave=int(whole)
    print("starting octave:",current_octave)
    first=0
    second=1
    third=2
    music_notes=""
    while third<len(decimals):
        try:
            octal=str(oct(int(decimals[first]+decimals[second]+decimals[third])))
        except:
            print(octal)
            pass
        print(octal)
        if '00' in octal and current_octave!=min_octave and current_octave<max_octave:
            current_octave =current_octave-1
        else:
            if '00' in octal and current_octave==min_octave:
                current_octave=current_octave+1
        if '7' in octal and current_octave<max_octave:
            current_octave =current_octave+1
        else :
            if '7' in octal and current_octave==max_octave:
                current_octave=2
        octal=octal.replace('00','')
        music_notes= music_notes+'C'+str(current_octave)+'-'
        octal=octal.replace('0o','')
        for number in octal:
            music_notes=music_notes+get_note(number,current_octave)
        
        #print(octal + ":" + music_notes)   
        first=first+3
        second=second+3
        third=third+3
        
    return music_notes

def get_note(number, current_octave):
    if number ==  '0':
        return 'C'+str(current_octave)+'-'
    if number == '1':
        return 'D'+str(current_octave)+'-'
    if number == '2':
        return 'E'+str(current_octave)+'-'
    if number == '3':
        return 'F'+str(current_octave)+'-'
    if number == '4':
        return 'G'+str(current_octave)+'-'
    if number == '5':
        return 'A'+str(current_octave)+'-'
    if number == '6':
        return 'B'+str(current_octave)+'-'
    return ''
```

```python
def get_wave(freq, duration=0.5):
    '''
    Function takes the "frequecy" and "time_duration" for a wave 
    as the input and returns a "numpy array" of values at all points 
    in time
    '''
    #print(freq)
    amplitude = 4096
    t = np.linspace(0, duration, int(samplerate * duration))
    wave = amplitude * np.sin(2 * np.pi * freq * t)
    
    return wave
```

```python
import midi
# Instantiate a MIDI Pattern (contains a list of tracks)
pattern = midi.Pattern()
# Instantiate a MIDI Track (contains a list of MIDI events)
track = midi.Track()
# Append the track to the pattern
pattern.append(track)
# Instantiate a MIDI note on event, append it to the track
on = midi.NoteOnEvent(tick=0, velocity=20, pitch=midi.G_3)
track.append(on)
# Instantiate a MIDI note off event, append it to the track
off = midi.NoteOffEvent(tick=100, pitch=midi.G_3)
track.append(off)
# Add the end of track event, append it to the track
eot = midi.EndOfTrackEvent(tick=1)
track.append(eot)
# Print out the pattern
print(pattern)
# Save the pattern to disk
midi.write_midifile("example.mid", pattern)
```
